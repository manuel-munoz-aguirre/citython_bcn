```python
### Libraries: 

#import urllib, json
import urllib.request
import json
import pandas as pd
```

```python
api_key = ""
```

```python
#Grabbing and parsing the JSON data

def GoogPlac(lat,lng,radius,types,key):
    #making the url
    AUTH_KEY = key
    LOCATION = str(lat) + "," + str(lng)
    RADIUS = radius
    TYPES = types
    MyUrl = ('https://maps.googleapis.com/maps/api/place/nearbysearch/json'
           '?location=%s'
           '&radius=%s'
           '&types=%s'
           '&sensor=false&key=%s') % (LOCATION, RADIUS, TYPES, AUTH_KEY)
    
    #grabbing the JSON result
    response = urllib.request.urlopen(MyUrl)
    jsonRaw = response.read()
    jsonData = json.loads(jsonRaw)
    
    return jsonData
```

```python
#Maximum radious = 50,000
#Location of Plaza Cataluya

a = GoogPlac(lat="41.386903", lng = "2.169404", radius = "500", types = "restaurant", key = api_key)
```
